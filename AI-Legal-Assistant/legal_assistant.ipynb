```python
import fitz
import streamlit
import langchain
import faiss

print("✅ All main packages installed successfully!")
```

```python
import os

pdf_path = "data/sample_contract.pdf"

if os.path.exists(pdf_path):
    print(" PDF found successfully!")
else:
    print(" PDF not found!")
    
```

```python
import fitz

# Open the PDF
pdf = fitz.open(pdf_path)

print("📄 Total pages:", len(pdf))

# Extract text from all pages
pages = []

for page_number, page in enumerate(pdf, start=1):
    text = page.get_text("text")

    if text.strip():
        pages.append({
            "page": page_number,
            "text": text
        })

pdf.close()

print("✅ Pages extracted:", len(pages))
```

```python
print("========== PAGE 1 ==========")
print(pages[0]["text"])
```

```python
from langchain_core.documents import Document

documents = []

for page in pages:
    document = Document(
        page_content=page["text"],
        metadata={
            "source": "sample_contract.pdf",
            "page": page["page"]
        }
    )

    documents.append(document)

print("✅ LangChain documents created:", len(documents))
```

```python
print("DOCUMENT TEXT:")
print(documents[0].page_content[:500])

print("\nMETADATA:")
print(documents[0].metadata)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)

chunks = text_splitter.split_documents(documents)

print(" Original pages:", len(documents))
print(" Total chunks created:", len(chunks))
```

```python
from langchain_huggingface import HuggingFaceEmbeddings

embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

print(" Embedding model loaded successfully!")
```

```python
test_vector = embedding_model.embed_query(
    "What are the termination conditions?"
)

print("✅ Embedding created successfully!")
print("Vector dimensions:", len(test_vector))
print("First 10 values:", test_vector[:10])
```

```python
from langchain_community.vectorstores import FAISS

vector_store = FAISS.from_documents(
    documents=chunks,
    embedding=embedding_model
)

print("✅ FAISS vector database created successfully!")
```

```python
question = "How can this agreement be terminated?"

results = vector_store.similarity_search(
    question,
    k=2
)

print("🔍 Search results:", len(results))

for i, result in enumerate(results, start=1):
    print(f"\n========== RESULT {i} ==========")
    print("📄 Source:", result.metadata.get("source"))
    print("📖 Page:", result.metadata.get("page"))
    print("\n", result.page_content)
```

```python
def ask_legal_assistant(question, k=3):

    # 1. Search FAISS for relevant legal text
    results = vector_store.similarity_search(
        question,
        k=k
    )

    # 2. Build context from retrieved chunks
    context_parts = []

    for result in results:
        source = result.metadata.get("source", "Unknown")
        page = result.metadata.get("page", "Unknown")

        context_parts.append(
            f"""
SOURCE: {source}
PAGE: {page}

{result.page_content}
"""
        )

    context = "\n\n".join(context_parts)

    # 3. Create the RAG prompt
    prompt = f"""
You are an AI Legal Document Assistant.

Answer the user's question using ONLY the document context provided below.

RULES:
1. Do not invent information.
2. If the answer is not in the context, say:
   "I could not find this information in the uploaded document."
3. Explain the answer clearly and simply.
4. Mention the relevant source file and page number.
5. Do not claim to be a lawyer.
6. State that the answer is for informational purposes and is not professional legal advice.

DOCUMENT CONTEXT:

{context}

USER QUESTION:

{question}

ANSWER:
"""

    # 4. Send retrieved context to Gemini
    response = client.models.generate_content(
        model="gemini-3-flash-preview",
        contents=prompt
    )

    return response.text, results
```

```python
question = "How can this agreement be terminated?"

answer, sources = ask_legal_assistant(question)

print("⚖️ AI LEGAL ASSISTANT")
print("=" * 60)

print("\nQUESTION:")
print(question)

print("\nANSWER:")
print(answer)
```
